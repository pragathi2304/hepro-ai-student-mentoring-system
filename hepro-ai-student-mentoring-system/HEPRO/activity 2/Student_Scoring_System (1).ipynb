# Activity 2: Student Scoring System (Rule-Based Intelligence)

This notebook computes APS, WWS, PTMS, CRS, and SRI from students.csv.

```

import pandas as pd

# Load dataset
df = pd.read_csv("students.csv")
df.head()
```

## Academic Performance Score (APS)

```

df["APS"] = (
    (df["gpa"] / 10) * 50 +
    (df["attendance"] / 100) * 30 +
    (df["assignments_completion"] / 100) * 20
)
```

## Wellness & Wellbeing Score (WWS)

```

df["sleep_score"] = df["sleep_hours"].apply(
    lambda x: 100 if 7 <= x <= 8 else max(0, 100 - abs(x - 7.5) * 20)
)
df["stress_score"] = (10 - df["stress_level"]) * 10

df["WWS"] = 0.6 * df["sleep_score"] + 0.4 * df["stress_score"]
```

## Productivity & Time Management Score (PTMS)

```

df["PTMS"] = (
    0.5 * df["productivity_score"] +
    0.3 * (100 - df["distractions"] * 10) +
    0.2 * df["engagement_score"]
)
```

## Career Readiness Score (CRS)

```

df["CRS"] = (
    0.4 * df["career_clarity"] +
    0.4 * df["skill_readiness"] +
    0.2 * df["engagement_score"]
)
```

## Student Readiness Index (SRI) and Classification

```

df["SRI"] = (
    0.30 * df["APS"] +
    0.25 * df["WWS"] +
    0.20 * df["PTMS"] +
    0.25 * df["CRS"]
)

def classify(sri):
    if sri >= 80:
        return "Green"
    elif sri >= 65:
        return "Blue"
    elif sri >= 50:
        return "Yellow"
    else:
        return "Red"

df["Category"] = df["SRI"].apply(classify)
df.head()
```

## Save Scored Dataset

```

df.to_csv("students_scored.csv", index=False)
print("Saved students_scored.csv")
```
